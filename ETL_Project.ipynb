```python
import pandas as pd
from sqlalchemy import create_engine
from config import user, password
```

```python
csv_file = "Resources/Chicago.csv"
Chicago_data_df = pd.read_csv(csv_file)
Chicago_data_df.head()
```

```python
new_Chicago_data_df = Chicago_data_df[['Inspection_ID', 'DBA_Name', 'Facility_Type','Risk','City','Inspection_Date','Inspection_Type','Results','Violations']].copy()
new_Chicago_data_df.head()
```

```python
new_Chicago_data_df['Inspection_Date']=pd.to_datetime(new_Chicago_data_df['Inspection_Date'])
new_Chicago_data_df.head()
```

```python
new_Chicago_data_df['Inspection_Date_Chicago']=new_Chicago_data_df['Inspection_Date'] 
new_Chicago_data_df['Inspection_Type_Chicago']=new_Chicago_data_df['Inspection_Type'] 

del new_Chicago_data_df['Inspection_Date'] 
del new_Chicago_data_df['Inspection_Type'] 
new_Chicago_data_df.head()
```

```python
csv_file_2 = "Resources/SF.csv"
SF_data_df = pd.read_csv(csv_file_2)
SF_data_df.head()
```

```python
new_SF_data_df = SF_data_df[['business_id', 'business_name', 'business_city','inspection_id','inspection_date','inspection_score','inspection_type','violation_id','violation_description','risk_category']].copy()
new_SF_data_df.head()
```

```python
new_SF_data_df['inspection_date']=pd.to_datetime(new_SF_data_df['inspection_date'])
new_SF_data_df.head()
```

```python
rds_connection_string = user + ":" + password + "@127.0.0.1/restaurant1_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_Chicago_data_df.to_sql(name='chicago_violation', con=engine, if_exists='append', index=False)
```

```python
new_SF_data_df.to_sql(name='sf_violation', con=engine, if_exists='append', index=False)
```
